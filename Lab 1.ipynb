```python
articleContent = []
for item in articles:
    articleContent.append(item.find('div',class_ ="entry-content"))
```

```python
for item in articleContent:
    print(item)
    print('------')
```

```python
import re

def clean_text(text):

    html = re.compile('[<,#*?>]')
    text = html.sub(r'',text)
 
    text = re.sub(r'&\w+;', '', text)

    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
pureArticles = []
for item in articleContent:
    pureArticles.append(clean_text(item.get_text()))
print(pureArticles[0])
```

```python
print(pureArticles[1])
```

```python
documents = []  # Initialize an empty list to store dictionaries

i = 1
for item in pureArticles:
    document = {"title": i, "content": item}
    documents.append(document)
    i += 1

print(documents)

```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017')
db = client.ArabicNewsletter

collection = db.Articles
collection.insert_many(documents)
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
arabic_stopwords = set(stopwords.words('arabic'))
documents = collection.find()
doc =  []
for document in documents:
    #Tokenization
    tokenized_content = word_tokenize(document['content'])
    #Stop words
    filtered_content = [word for word in tokenized_content if word not in arabic_stopwords]
    #Normalization
    normalized_content = [word.lower() for word in filtered_content]
    document['content'] = normalized_content
    doc.append(document)
```

```python
#Stemming
from nltk.stem import PorterStemmer
doc2 = doc
stemmer = PorterStemmer()
for document in doc:
    stemmed_content = [stemmer.stem(word) for word in filtered_content]
    document['content'] = stemmed_content
    print(document)
    
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for document in doc2:
    lemmatized_content = [lemmatizer.lemmatize(word) for word in filtered_content]
    document['content'] = lemmatized_content
    print(document)
```

##### Stemming cuts off prefixes and suffixes to find the root form of words, while lemmatization considers context and meaning to return the base or dictionary form. Stemming is faster but less accurate, while lemmatization is slower but more accurate.

```python
# Parts of Speech (POS)
from nltk import pos_tag
for document in doc2:
    tagged = pos_tag(normalized_content)
    print(tagged)
```

```python
# Named Entity Recognition(NER)
from nltk import  ne_chunk

named_entities = ne_chunk(tagged)
print(named_entities)
```
